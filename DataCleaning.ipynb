```python
# Creamos una función que se repeterirá para obtener los datos
def filtrarDatos(filtro, df):
    """
    Esta función recibe como parámetro una tupla filtro (provincia, municipio, estación)
    y un df.
    Devuelve un df filtrado.
    """
    df = df[(df["PROVINCIA"] == filtro[0]) & (df["MUNICIPIO"] == filtro[1])]
    #Mediante el uso de la serie, selecciono unos datos específicos del df
    return df
```

```python
# Limpieza de datos
print("Iniciando limpieza...")
for anyo in range(2009, 2021): #Iteración sobre todos los años
    path = "./Datos " + str(anyo) + "/Horarios/" #Dirección relativa de los ficheros de datos 
    print("Limpiando año " + str(anyo))
    for filename in totalDatos[anyo]: #Iteración sobre los compuestos de un año
        df_inicial = pd.read_excel(path + filename + ".xlsx") #Obtengo el df del compuesto del año correspondiente
        aux = filename.split("_")
        compuesto = aux[0]
        for lugar, condiciones in lugares.items():
            try:
                df = filtrarDatos(condiciones, df_inicial) #Filtro el df
                if not df.empty: #Si no está vacío, lo guardo y añado una tupla al resumen
                    if anyo in resumen:
                        resumen[anyo].append((lugar, compuesto, True))
                    else:
                        resumen[anyo] = [(lugar, compuesto, True)]
                    print(f"Introduciendo datos de {lugar} en {compuesto}_{anyo}.xlsx")
                    writer = pd.ExcelWriter(compuesto+"_"+str(anyo)+".xlsx", engine='openpyxl', mode='a', if_sheet_exists='overlay')
                    
                    #Tenemos un archivo "compuesto_anyo.xlsx" para cada compuesto y su año, posteriormente los pegaremos todos en un
                    #archivo denominado "compuesto.xlsx" con los datos de cada año en una hoja.
                    #Se utiliza esto para evitar sobrecargar el objeto writer con muchos datos en su variable books
                    
                    writer.book = load_workbook(compuesto+"_"+str(anyo)+".xlsx")
                    writer.sheets =  {ws.title: ws for ws in writer.book.worksheets}
                    df.to_excel(writer, sheet_name=str(anyo), startrow=writer.sheets[str(anyo)].max_row, index=False, header=False)
                    writer.close()
                    writer.handles = None
                    print("Datos introducidos")
                    print("\n")
                else: #Si está vacío, no introduzco nada 
                    print("El df está vacío")
                    print(f"No se introdujo nada de {lugar} en {compuesto}_{anyo}.xlsx")
                    print("\n")
                    if anyo in resumen:
                        resumen[anyo].append((lugar, compuesto+"_"+str(anyo)+".xlsx", False))
                    else:
                        resumen[anyo] = [(lugar, compuesto+"_"+str(anyo)+".xlsx", False)]

            except Exception as e:
                print(f"No se pudo introducir los datos de {lugar} en {compuesto}_{anyo}.xlsx")
                print("\n")
                print(f"El error es:\n{e}")
                if anyo in resumen:
                    resumen[anyo].append((lugar, compuesto+"_"+str(anyo)+".xlsx", False))
                else:
                    resumen[anyo] = [(lugar, compuesto+"_"+str(anyo)+".xlsx", False)]
                    
print("Limpieza completada")
```
