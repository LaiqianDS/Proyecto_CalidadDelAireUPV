```python
import pandas as pd
```

```python
#We are going to calculate the Air Quality Index for each chemical
#We have to remember that each one has diferentel values for each level

def calculaICA_so2(valor):
    try:
        valor = float(valor)
        if 0 <= valor <= 100:
            return 1
        elif 100 < valor <= 200:
            return 2
        elif 200 < valor <= 350:
            return 3
        elif 350 < valor <= 500:
            return 4
        elif 500 < valor <= 750:
            return 5
        elif 750 < valor <= 1250:
            return 6
        else: return 0
    except:
        return 0
    
def calculaICA_pm25(valor):
    try:
        valor = float(valor)
        if 0 <= valor <= 10:
            return 1
        elif 10 < valor <= 20:
            return 2
        elif 20 < valor <= 25:
            return 3
        elif 25 < valor <= 50:
            return 4
        elif 50 < valor <= 75:
            return 5
        elif 75 < valor <= 800:
            return 6
        else: return 0
    except:
        return 0

def calculaICA_pm10(valor):
    try:
        valor = float(valor)
        if 0 <= valor <= 20:
            return 1
        elif 20 < valor <= 40:
            return 2
        elif 40 < valor <= 50:
            return 3
        elif 50 < valor <= 100:
            return 4
        elif 100 < valor <= 150:
            return 5
        elif 150 < valor <= 1200:
            return 6
        else: return 0
    except: 
        return 0

def calculaICA_o3(valor):
    try:
        valor = float(valor)
        if 0 <= valor <= 50:
            return 1
        elif 50 < valor <= 100:
            return 2
        elif 100 < valor <= 130:
            return 3
        elif 130 < valor <= 240:
            return 4
        elif 240 < valor <= 380:
            return 5
        elif 380 < valor <= 800:
            return 6
        else: return 0
    except:
        return 0
    
def calculaICA_no2(valor):
    try:
        valor = float(valor)
        if 0 <= valor <= 40:
            return 1
        elif 40 < valor <= 90:
            return 2
        elif 90 < valor <= 120:
            return 3
        elif 120 < valor <= 230:
            return 4
        elif 230 < valor <= 340:
            return 5
        elif 340 < valor <= 1000:
            return 6
        else: return 0
    except: return 0
```

```python
filename = "DatosFinales.xlsx"
df = pd.read_excel(filename, sheet_name="Completo")

col_name = list(df.columns)
places = col_name[2:] #We get all the places name
```

```python
#Each place has the format "place/ chemical"
#We are going to calculate value for the places whose chemical is measured in de AQI

so2_places = [place for place in places if "SO2" in place]
pm25_places = [place for place in places if "PM25" in place]
pm10_places = [place for place in places if "PM10" in place]
o3_places = [place for place in places if "O3" in place]
no2_places = [place for place in places if "NO2" in place]
```

```python
#Now, we are going to apply the map() function to get the correspondence and store it in a new column
#Threshold values are in the Spanish National Air Quality Index

for place in so2_places:
    df[place+" ICA"] = df[place].map(calculaICA_so2)
for place in pm25_places:
    df[place+" ICA"] = df[place].map(calculaICA_pm25)
for place in pm10_places:
    df[place+" ICA"] = df[place].map(calculaICA_pm10)
for place in o3_places:
    df[place+" ICA"] = df[place].map(calculaICA_o3)
for place in no2_places:
    df[place+" ICA"] = df[place].map(calculaICA_no2)
```

```python
#Now, we are going to create a new column that shows the AQI for each place
#We obtain the value by picking the worst level of AQI of the place

aqi_values = {
    1 : "Buena",
    2 : "Razonablemente buena",
    3 : "Regular",
    4 : "Desfavorable",
    5 : "Muy desfavorable",
    6 : "Extremadamante desfavorable",
    0 : "Dato no válido"
}

chemicals = ["SO2", "NO2", "PM10", "PM25", "O3"]
unique_places = [place[:-5].strip() for place in so2_places]
place_chemicals = {}
for place in unique_places:
    place_chemicals[place] = []
    for chemical in chemicals:
        place_chemicals[place].append(place+"/ "+chemical +" ICA")

for place, lista in place_chemicals.items():
    df[place + " ICA"] = df[[lista[0], lista[1], lista[2], lista[3], lista[4]]].max(axis=1).map(aqi_values)
    df[place + " ICA num"] = df[[lista[0], lista[1], lista[2], lista[3], lista[4]]].max(axis=1)

df.head()
```

```python
df.to_excel('DatosICA.xlsx', index=False)
```
